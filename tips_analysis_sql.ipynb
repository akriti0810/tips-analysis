```python
import sqlite3
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
```

```python
conn = sqlite3.connect(":memory:")
df.to_sql("tips", conn, index=False)
```

```python
query = "SELECT * FROM tips LIMIT 5"
pd.read_sql(query, conn)
```

```python
query = "SELECT * FROM tips WHERE day = 'Sun'"
pd.read_sql(query, conn)
```

```python
query = """
SELECT day, AVG(tip / total_bill * 100) as avg_tip_pct
FROM tips
GROUP BY day
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT day, AVG(tip / total_bill * 100) as avg_tip_pct
FROM tips
GROUP BY day
ORDER BY avg_tip_pct DESC
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT day, AVG(tip / total_bill * 100) as avg_tip_pct
FROM tips
GROUP BY day
HAVING avg_tip_pct > 16
ORDER BY avg_tip_pct DESC
"""
pd.read_sql(query, conn)
```
